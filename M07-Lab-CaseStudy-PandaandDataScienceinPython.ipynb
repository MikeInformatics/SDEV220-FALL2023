Aggregate Statistics (Group By)

```python
salaries = pd.read_csv('Salaries.csv')

salaries.groupby(['Year'])
```

Working with Large Amount of Data

```python
new_salaries = pd.DataFrame(columns=salaries.columns)

for salaries in pd.read_csv('Salaries.csv', chunksize=5):
    results = salaries.groupby(['JobTitle']).count()
    
    new_salaries = pd.concat([new_salaries, results])
```

Plotting using matplotlib

```python
salaries = pd.read_csv('Salaries.csv')
salaries.plot()
plt.show()
```
